## Common r.v.
### Descrete
**Expected Value** for discrete r.v.:
$$E[X] = \sum_k x_k Pr[X=x_k]$$
$$\text{Note: }\underbrace{Pr[X=x_k]}_\text{pmf}$$
- Always deterministic (constant)
- The expectation operation 'squeezes' the randomness out of an r.v.
- Expectation is a linear operation (i.e. $f(ax) = af(x)$ and $f(a+b)=f(a)+f(b))$ 

**Expected Value** for continuous r.v.:
$$E[X] = \int_{k=-\infty}^{k=\infty} x_k f(x) dx$$


**Variance** of a r.v.(simplification of the variance expression):
$$Var[X]\sum_k E[(X_k - E[X])^2] = E[X^2+(E(x))^2 - 2XE[X]]$$
$$= E[X^2]+E[\underbrace{(E(x))^2}_\text{constant squared}] - E[\underbrace{2}_\text{constant}X\underbrace{E[X]}_\text{constant}]$$
$$= E[X^2]+(E(x))^2 - 2E[X]E[X]$$
$$= E[X^2]+(E(x))^2 - 2(E(x))^2$$
$$= E[X^2]-(E(x))^2$$




- The expectation operation 'squeezes' the randomness out of an r.v. - what was a r.v. which had a pdf, cdf, etc, comes out as just a number
- Both $E(x)$ and $Var(x)$ are numbers.
    - Deterministic (constant)
    - No longer an r.v.
    - In contrast, if we take f(x), it is still an r.v.

### Bernoulli
e.g. Coin toss
$$\Omega = \{H, T\}$$
$$= \{\underbrace{H}_\text{1}, \underbrace{T}_\text{0}\}$$
$$= \{\underbrace{H}_\text{p}, \underbrace{T}_\text{1-p}\}$$
- p is a parameter to the distribution function
- Expected value:
$$E[X]=1\cdot p+ 0\cdot (1-p) = p$$
- Variance:
$$E[(X - E[X])^2]$$

$$E[f(x)]=\sum_k f(x_k)Pr[X=X_k]$$
$$E[X^2]=\sum_k x^2_kPr[X=X_k]$$
$$E[f(x)+g(x)]=\sum_k (f(x_k) + g(x_k))\cdot p_k$$
$$=\sum_k f(x_k)Pr[X=x_k] + \sum_k g(x_k)Pr[X=x_k]$$


- Expected value of a constant:
$$E[c]=\sum_k c \cdot P_k = c \cdot \underbrace{\sum_k p_k}_\text{defined as 1} = c$$

### Geometric r.v.
"How many tosses *until* I see head (success). Assumes iid.
$$\Omega=\{1,2,...,\infty\}$$
$$P_k=Pr[X=k]$$
$$P_k=Pr[k-1 \text{ tails & }1 \text{ head}]$$
$$=\underbrace{(1-p)(1-p)...(1-p)}_\text{k-1 times}\cdot p$$
$$\boxed{p_k=(1-p)^{k-1}\cdot p}$$

- Note we multiply conjunction of independent event - $(1-p)\cdot (1-p)\cdot ...\cdot (1-p)\cdot p$
- Single parameter, $p$
- $k$ is not a parameter. Given a $k$, # of success, we can compute the probability

#### Expected value of Geometric r.v.
$$E[X]=\sum_{k=1}^{\infty}\underbrace {k\cdot(1-p)^{k-1}}_{-\frac{d}{dp}[(1-p)^k]} \cdot p$$
$$=p\sum_{k=1}^{\infty}\underbrace {k(1-p)^{k-1}}_{-\frac{d}{dp}[(1-p)^k]}$$
$$=p\sum_{k=1}^{\infty}-\frac{d}{dp}[(1-p)^k]$$
$$=-p\sum_{k=1}^{\infty}\frac{d}{dp}[(1-p)^k]$$
$$=-p\frac{d}{dp}\sum_{k=1}^{\infty}[(1-p)^k]$$
Sum of derivative is the derivative of sum (linear property)
$$=-p\frac{d}{dp} \left[\frac{1}{1-(1-p)} \right]$$
$$=-p\frac{d}{dp} \left[\frac{1}{p} \right]$$
$$=-p\frac{d}{dp} \left[p^{-1} \right]$$
$$=-p\cdot -p^{-2} $$
$$=\frac{p}{p^2}$$
$$=\boxed{\frac{1}{p}}$$

#### Variance of Geometric r.v.
$$=\boxed{\frac{1-p}{p^2}}$$


### Binomial r.v.

$S = {0,1,...,n} $, iid

n coin osses in parallel, how many heads come up?

$$Pr[X=k] = {n\choose k} p^k(1-p)^{n-k}$$
$$= \underbrace{{n\choose k}}_\text{how many ways k out of n coins are heads} \cdot p^k \cdot (1-p)^{n-k}$$
Binomial$(n,k,p)$

#### Expected value of Binomial r.v.
Note each outcome of the iteration of summation is an outcome of bernoulli(p).
$$Binomial(n,k,p) = \sum_{k=1}^{n}{n\choose k} \underbrace{p^k(1-p)^{n-k}}_{Bernoulli(p)}$$
Each bernoulli trials have the same weight/probability (e.g. coin toss is a coin toss, the probability of the coin for head/tail remains the same).

$$E[X] = E[ \underbrace{\text{ Bernoulli }(p) + \text{ Bernoulli }(p) + ... + \text{ Bernoulli }(p)}_\text{ n times} ]$$
$$ = \underbrace{E[\text{ Bernoulli }(p)] + E[\text{ Bernoulli }(p)] + ... + E[\text{ Bernoulli }(p)]}_\text{ n times}$$
$$ = \underbrace{p + p + ... + p}_\text{ n times}$$
$$=\boxed{np}$$

#### Variance of Binomial r.v.
$$Var(X) = Var( \underbrace{\text{ Bernoulli }(p) + \text{ Bernoulli }(p) + ... + \text{ Bernoulli }(p)}_\text{ n times} )$$
$$ = \underbrace{Var(\text{ Bernoulli }(p)) + Var(\text{ Bernoulli }(p)) + ... + Var(\text{ Bernoulli }(p))}_\text{ n times}$$
$$ = \underbrace{p\dot(1-p) + p\dot(1-p) + ... + p\dot(1-p)}_\text{ n times}$$
$$=\boxed{np\dot(1-p)}$$

## Expected value and variance of indicator r.v.

**Expected Value** of indicator r.v.:
$$\int_{\Omega} I_A(\omega) \cdot \underbrace{f_x(\omega)d\omega}_{dp} = $\int_{\Omega}I_A(\omega)dp$$
$$=\int_{A}I_A(x)dp + \int_{A^c}I_A(x)dp$$
$$=\underbrace{\int_{A}I_A(x)dp}_{=1} + \underbrace{\int_{A^c}I_A(x)dp}_{=0}$$
$$=\int_{A}1dp$$
$$=\int_{A}dp$$
$$=Pr(A)$$


**Variance** of an indicator r.v.:
$$Var(X)=\int_{\Omega} (I_A(x) - \underbrace{E[I_A]}_{Pr(A)})^2 dp$$
$$=\int_{\Omega} (I_A(x) - Pr(A))^2 dp$$
$$ = \int_{\Omega} \left (I_A(x)^2 + Pr(A)^2 - 2I_A(x)\cdot Pr(A) \right )dp$$
Since $I_A = 1$, $I_A^2 = 1$, $I_a = 0$, $I_a^2 = 0$,
$$ = \int_{\Omega} \left (I_A(x) + Pr(A)^2 - 2I_A(x)\cdot Pr(A) \right )dp$$

$$ = \int_{\Omega} I_A(x) dp + \int_{\Omega}Pr(A)^2 dp - \int_{\Omega}2I_A(x)\cdot Pr(A)dp$$
$$ = \underbrace{\int_{\Omega} I_A(x) dp}_{Pr(A)} + Pr(A)^2 \int_{\Omega}dp - 2Pr(A)\underbrace{\int_{\Omega}I_A(x)dp}_{Pr(A)}$$
$$ = Pr(A) + Pr(A)^2 \int_{\Omega}dp - 2Pr(A)\cdot Pr(A)$$
$$ = Pr(A) + Pr(A)^2 - 2Pr(A)^2$$
$$ = Pr(A) - Pr(A)^2$$
Since $A = {1,0}$, and $1^2 = 1$ and $0^2 =0$, it is equivalent to $Pr(A^2)- Pr(A)^2$
$$=E[I_a(A)^2] - (E[I_a])$$

## Proof: expected value of binomial r.v.
$$\require{cancel}$$  
$$E[X] = \sum_k x_k Pr[X=x_k]$$
$$= \sum_{k=0}^n k\cdot {n \choose k}p^k (1-p)^{n-k}$$
$$= \sum_{k=0}^n k\cdot \frac{n!}{(n-k)!k!} p^k(1-p)^{n-k}$$
Pull out n and p (the goal is to show the expression on the righ is equal to $np$, so we want the summation to evaluate to $1$:
$$= np\sum_{k=0}^n k\cdot \frac{n\cdot(n-1)!}{(n-k)!k!} p\cdot p^{k-1}(1-p)^{n-k}$$
Cancel out k:
$$= np\sum_{k=0}^n k \cdot \frac{n\cdot(n-1)!}{(n-k)!k\cdot (k-1)!} p\cdot p^{k-1}(1-p)^{n-k}$$
$$= np\sum_{k=0}^n {\cancel{k}} \cdot \frac{n\cdot(n-1)!}{(n-k)!{\cancel{k}}\cdot (k-1)!} p\cdot p^{k-1}(1-p)^{n-k}$$

$$= np\sum_{k=0}^n \frac{(n-1)!}{(n-k)!(k-1)!} p^{k-1}(1-p)^{n-k}$$
Express $n-k$ as $(n-1)-(k-1)$:
$$= np\sum_{k=0}^n  \frac{(n-1)!}{((n-1)-(k-1))!(k-1)!} p^{k-1}(1-p)^{(n-1)-(k-1)}$$
Change the starting k index to 1, since when k is $0$, the summond is 0, so it does not contribute to the sum anyways:
$$= np\sum_{k=1}^n \frac{(n-1)!}{((n-1)-(k-1))!(k-1)!} p^{k-1}(1-p)^{(n-1)-(k-1)}$$
Let $j = k-1$ and $m=n-1$, and substitute variables, noting when $k = 1$, $j=1-1=0$, and when $k=n$, $j=n-1$, which is m:

$$= np\sum_{j=0}^n\frac{m!}{(m-j)!j!} p^j(1-p)^{m-j}$$
$$= np \underbrace{\sum_{j=0}^n\frac{m!}{(m-j)!j!} p^j(1-p)^{m-j}}_\text{Sums to 1, according to the axiom of probability}$$

$$= np\cdot 1$$
$$= np$$
