## Linearity and Expectation

### Example 1
N students graduate. all toss their hats, and pick a random one. What is expected number of students who gets his/her own hat?

#### Indicator r.v. 

$
X_i=  \begin{cases} 
      1 \text{ if student picks up own hat} \\
      0 \text{ else }
   \end{cases}
$

What is the expected number of students who get their own hat?


**Computation will be very complex**

- No one picks own hat: $ (1 - \frac{1}{n})$
    - because $E[X_i] = \frac{1}{n}$...
- At least one person picks up own hat: $P[Y = 1]$
    - $Y$: number of students picking up their own hat
- Computation gets very complicated ...

**Expected value of sum of indicator random variables**

$E\left[\sum_{i=1}^n X_i\right]$ 

$Y = \sum_{i=1}^n X_i$  <- number of students picking up their own hat

**Expected value of a single indicator random variable**
$$E(X_i) = P(1) \cdot 1 + (1 - P) \cdot 0 = P$$


$$\sum k P_k$$
$$E[X_i] = p \cdot 1 + (1-p) \cdot 0$$
$$ = \frac {1}{n} \text { for } \forall i$$

**Linearity of expectation applied**

Expected value per each trial: $E\left[\sum X_i\right]$
$$Y = \sum E(X_i) = \sum_{i=1}^n \frac{1}{n} = n \cdot \frac{1}{n} = 1$$

**Linearity of expectation for a linear combination of random variables**

$$E[cX + dY] = cE[X] + dE[Y]$$

# Marcov's Inequality (M.I.)

$$P[X \geq a] \leq \frac{E[X]}{a}$$

How far off the outlieres can be is bounded.

```python
Image(filename='images/session5/markov-ineq.png', width=550)
```

Dashed line above is the complementary cumulative distribution function (CCDF)

$$E[X]=\int_{0}^{\infty} P[X \geq x] \text{ } dx$$

## Proof (M.I.)
$$E[X] = \int_{-\infty}^{\infty} P[X \geq x] \text{ }dx$$
$$ = \int_{-\infty}^{a} P[X \geq x] \text{ }dx + \int_{a}^{\infty} P[X \geq x] \text{ }dx$$
$$ = \int_{-\infty}^{a} xf_x(x) \text{ }dx + \int_{a}^{\infty} xf_x(x) \text{ }dx$$
$$\text{Since }\int_{a}^{\infty} xf_x(x) \text{ }dx\text{ is a subset of} \int_{-\infty}^{\infty} xf_x(x) \text{ }dx $$
$$ \geq \int_{a}^{\infty} xf_x(x) \text{ }dx$$
$$\text{Since a is the lower bound,}$$
$$ \geq \int_{a}^{\infty} af_x(x) \text{ }dx$$
$$ \geq a \int_{a}^{\infty} f_x(x) \text{ }dx $$
$$ \geq a \underbrace{\int_{a}^{\infty} f_x(x) \text{ }dx}_{P[X \geq x]} $$

$$ = a P[X \geq x]$$
So,

$$ E[X]\geq a P[X \geq x]$$
$$ \frac{E[X]}{a} \geq  P[X \geq x]$$
$$\boxed{P[X \geq x] \leq \frac{E[X]}{a} }\text{ for } \forall \text{ }a \in [0, \infty]$$

### Example

Say students have average score in exam $= 27$. What is the probability that a student has a score $\geq 90$?
#### M.I.
$$P[X \geq 90] \leq \frac{27}{90} $$
$$= \frac{3}{10} \approx 33\%$$

## Chebyshev's Inequality (C.I)

$$P[|X - E[X]| \geq a] \leq \frac{\text{Var}(X)}{a^2}$$
$$ \text{OR} $$
$$P[|X - E[X]| \geq a] \leq \frac{\sigma^2}{a^2}$$

## Proof (C.I.)
Let $D^2 = (X-E[X])^2$, note $(\underbrace{X}_\text{r.v.}-\underbrace{E[X]}_\text{constant})$

$$P\left[X \geq a\right] \leq \frac{E[X]}{a} \text{ ............. M.I.}$$
$$P\left[D^2 \geq a^2 \right] \leq \frac{E[D^2]}{a^2}$$
$$\leq \frac{Var[X]}{a^2} = \frac{\sigma^2}{a^2}$$
$$\therefore P\left[D^2 \geq a^2 \right] = \frac{\sigma^2}{a^2}$$
$$ P\left[\sqrt{D^2} \geq \sqrt{a^2} \right] = \frac{\sigma^2}{a^2}$$
$$ P\left[{D} \geq a \right] = \frac{\sigma^2}{a^2}$$
$$ P\left[{X-E[X]} \geq a \right] = \frac{\sigma^2}{a^2}$$




### Example

The average wait time for Uber is $10$ min., and $\sigma = 1$ min. What is the probability I have to wait more than $15$ min?

#### M.I.
$$P[X \geq 15] \leq \frac{10}{15} $$
$$= \frac{2}{3} \approx 66\%$$

#### C.I.
$$P[|X - E[X] > a|] = \frac{\sigma^2}{a}$$
We know X > 0, so
$$P[X - 10 > 15|] = \frac{1^2}{15}$$
$$P[X > 25] = \frac{1}{25} = 40\%$$

## Covariance
Covariance easures how 2 r.v.'s move together ($X\uparrow \rightarrow Y\downarrow, X\uparrow \rightarrow Y\uparrow, X\downarrow \rightarrow Y\downarrow$ etc)
$$\boxed{Cov(X,Y) = E\left[(E[X-E[X]])(E[Y-E[Y])\right]}$$
$$= E\left[(E[XY] - XE[Y] - E[X]Y + E[X]E[Y])\right]$$
$$= E[XY] - E[X]E[Y] - E[X]E[Y] + E[X]E[Y]$$
$$= E[XY] - E[X]E[Y] - {\cancel{E[X]E[Y]}} + {\cancel{E[X]E[Y]}}$$
$$= \boxed{E[XY] - E[X]E[Y]}$$


If X and Y are independent, then
$$f_{X,Y}(x,y)=f_X(x)f_Y(y)$$
$$E[XY]=\int_Y \int_X f_{XY}(x,y) \text{ }dxdy$$
$$=\int_X f_{X}(x) \text{ }dx\int_Y f_{Y}(y)\text{ }dy$$
$$= E[X]E[Y]$$

## Correlation
Correlation quantifies the strength and direction of the relationship between 2 r.v.'s, scaled. Correlation coefficient is normalized to be between $-1$ to $1$ (i.e. $-1 \leq \rho_{XY} \leq +1 $)
$$\underbrace{\rho_{XY}}_\text{Correlation Coefficient} = \frac{\sigma_{XY}}{\sigma_X \sigma_Y}$$
$$= \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X) \text{Var}(Y)}}$$

- Highest corr/cov when $X=Y$
- Lowest corr/cov when $X=-Y$

**Notes on Covariance, Correlation, and Independence**

- Independence $\rightarrow 0$ Cov $\rightarrow 0$ correlation is true.
- But $0$ Cov/Correlation $\nRightarrow$ independence, in general
- Cov is $0 \iff E[XY] = E[X]E[Y]$

**Covariance and correlation measures linear relationship well, but not so much for non-linear relationship**

### Example
$
X:  \begin{cases} 
      +1 \text{ ... with probability } 0.5 \\
      -1 \text{ ... } 0.5 \\
   \end{cases}
$

$
Y=  \begin{cases} 
      0 \text{ if } X=-1 \\
      -1 \text{ or }+1  \text{ with probability } 0.5 \text{ if } X=1
   \end{cases}
$

$Cov(XY) = E[ (X - E(X)) (Y - E(Y)) ]$

$      = 0$

- $\Omega = \{(1, -1), (1, 1), (-1, 0)\}$
- Computing $\sum_{i=1}^{n} \sum_{j=1}^{m} x_i y_jP[X=x_i, Y=y_j]$:
    - $E[XY] = +/-0 * P(X=-1) = 0$ 
    -        $+ 1*1 * P(X=1, Y=1)$
    -       $+ 1*(-1) * P(X=1, Y=-1)$
    -       $= 0$
    - $\therefore$ their covariance is $0$
    
Note: $n=2$ and $m=3$ for the bound of the sums, $\sum_{i=1}^{n} \sum_{j=1}^{m} x_i y_jP[X=x_i, Y=y_j]$ because $S_X = \{+1, -1\}$ and $S_Y = \{0, +1, -1\}$.

**$\therefore$ covarience is zero, but X and Y are clearly dependent (i.e. Y's outcome depends on X's outcome).**

#### Remarks
- Covariance and correlation only talks about the behavior of the product of expectations
- Independence is tied to the whole probability distribution of the joint distribution
- $\therefore$ independence is a stronger statement

## Jointly Gaussian (JG, Correlatedness and independence)
A joint distribution is JG when r.v. X and Y are both normally distributed

Ugly expression, but when plot their contour in 2D, it tends to elipse where its center corresponds to the mean of the distribution

#### Joint Distribution
$$f_{X,Y}(x,y) = \frac{ \text{exp}\left[ -\frac{1}{2(1-\rho^2)}  \left(\frac{(x-\mu_X)^2}{\sigma_X^2} - \frac{2\rho (x-\mu_X)^2(y-\mu_Y)^2 (y-\mu_Y)^2(y-\mu_Y)^2}{\sigma_X \sigma_Y} + \frac{(y-\mu_Y)^2}{\sigma_Y^2}  \right) \right] }{ 2 \pi \sigma_1 \sigma_2 \sqrt{1 + \rho_{XY}^2} }$$

#### pdf
$$f_{\underline{X}}(\underline{x}) = \frac{\text{exp}\left[-\frac{1}{2}(\underline{x}-\underline{\mu})^TK^{-1}(x-\underline{\mu})\right]}{\left(2 \pi \right)^{\frac{n}{2}} |K|^{\frac{1}{2}}}$$
#### Note:
- Uncorrelated + JG $\Rightarrow$ independence

#### Covariance Matrix
$
K_1 = \begin{bmatrix}
\sigma_X^2 & \text{Cov}(X,Y) \\
\text{Cov}(X,Y) & \sigma_Y^2
\end{bmatrix}
$

$
K_2 = \begin{bmatrix}
\\ \text{Var}(X_1) & \text{Cov}(X_1,Y_2) & \text{Cov}(X_1,Y_3)\\
\\ \text{Cov}(X_2,Y_1) & \text{Var}(X_2) & \text{Cov}(X_2,Y_3)\\
\\ \text{Cov}(X_3,Y_1) &  \text{Cov}(X_3,Y_2) &  \text{Var}(X_3)\\
\end{bmatrix}
$

Note: $\text{Var}(X) = \sigma^2_X$

#### A Vector of Expected Values

K_2 = \begin{bmatrix}

$
