If $X_i$'s are independent:
$$\text{Var}\left(\sum X_i \right) = n \text{Var}(x)$$

$$\text{Var}\left(a X\right) = a^2 \text{Var}(X)$$
together, you get $\text{Var}(\frac{\sum Xi}{n}) = \frac{1}{n^2} \cdot n \cdot \text{Var}(x)$
which is $\text{Var}(X_n)$

For any r.v. $X$ with mean $\mu$ and standard diviation $S$ ,   $Y = \frac{(X-u)}{S}$ is a normalized version of that r.v.

For continuous r.v., variance is expressed as
 $$\text{Var}(X) = \int x^2 f(x) \text{ dx}$$ 
 $$= \int (x-\mu)^2 f(x) \text{ dx}$$

### Characteristic Functions
https://sas.uwaterloo.ca/~dlmcleis/s901/chapt6.pdf
- Characteristic function is called the moment-generating function (mgf) of an r.v.
- Similarly to Laplace and Fourier transform (e.g. Fourier transform takes a data from time domain to frequency domain), mgf allows us to take an r.v., and transform it to the t domain ("moment" domain)
    - Characterize the data uniquely
    - Allows easier mathematical manipulation of data (e.g. Fourier transform allows convolution in the time domain to multiplication in frequency domain)

```python
# https://en.wikipedia.org/wiki/Moment-generating_function
Image(filename='images/session8/properties_of_mgf.png', width=250)
```

$$M(t) = E[e^{tX}]$$
$$=\int e^{tX} f(x) \text{ dx}$$
- X: r.v.
- t: some number (a free variable that allows us to take the data to a different domain)
- An analogy might be we take the signal in time domain to frequency domain (f is analogous to g)
- mgf is unique to the distribution
   - Each distribution has its own unique pdf, cdf, and mgf
   
- To evaluate $E[X^n]$, given $M(t)$,

$$E[X^n] = \frac{d^n}{d^nt} M_x(t) \bigg\rvert_{t=0}$$
   
## Applications
### 1)
- Easier calculations of sums of r.v. - sum operation becomes easier to do
    - take two M(t), multiplying becomes sum ($e^{tX} \cdot e^{tY} = e^{t(X + Y)})$
- If I have mgf, we can reverse engineer/ take the inverse, and get pdf/cdf etc

### 2)

## Conditional Entropy
$$H(Y|X) = \sum_X p(x) H(Y | X=x)$$
$$= -\sum_X p(x) \sum_Y p(y|x) \log_2 p(Y | X)$$
$$= -\sum_X \sum_Y p(y|x)  p(x) \log_2 p(Y | X)$$
$$= -\sum_X \sum_Y p(x,y) \log_2 p(Y | X)$$

When considering the sum of independent random variables $X_1, X_2, \ldots, X_n$, the standard deviation of their sum $S = \sum_{k=1}^n X_k$ can be expressed as:
