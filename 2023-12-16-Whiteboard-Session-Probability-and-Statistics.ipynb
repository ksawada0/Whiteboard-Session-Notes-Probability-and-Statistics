## Axioms of Probability (For Finite Sample Space)

1. $P(S) = 1$

## Exclusive Events
$$A \cap B = \varnothing$$
- If $A \cap B = \varnothing$, then A and B are disjoint or exclusive, and $P(A \cup B) = P(A) \text{ and } P(B)$ holds.
- Exclusive Events = Disjoint Events
- if $A \cap B=\varnothing$, intuitively, what should $P(B|A)$ be?
    - Recall,
    $$P(B|A) = \frac{P(A \cap B)}{P(A)}$$
$$P(B|A)=\boxed{\varnothing}$$
- Given that I know for sure some $x \in B$ occurred,
$$\frac{P(\varnothing)}{P(A)}=0$$

## Independent Events
Two events A and B are independent iff $$P(A \cap B)=P(A)P(B)$$

- Note: This implies 
$$ P(B|A)=\frac{P(B|A)}{P(A)}=\frac{  P(A)P(B)}{P(A)}$$
$$ P(B|A)=\frac{P(B|A)}{P(A)}=P(B)$$
- Intuitively, knowing A occured does not change the likelihood of B's occurrance - A does ont affect B

1. We proved if A, B are disjoint, then $P(B|A)=\varnothing$
1. We showed if A, B are independent, then $P(B|A)=P(B)$
1. If $P(B|A)=\varnothing$, then it cannot be independent. So disjoint/exclusive event cannot be independent, and vice versa

## Combinatorics
### Experiment 1
Draw three cards from a deck. What is the probability we get 3 A's?

- Note each outcome is equally likely
- Here, order does NOT matter

$$P(A) = \frac{\text{# of desired outcomes in given event}}{\text{# of all possible outcomes}}$$
$$ = \text{# of desired outcomes in given event}(\underbrace{\frac{1}{\text{# of all possible outcomes}}}_\text{equally likely probability of each})$$
$$=\frac{4}{{52 \choose 3}}$$
